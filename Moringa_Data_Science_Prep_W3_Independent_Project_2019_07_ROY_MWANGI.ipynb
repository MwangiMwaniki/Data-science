WORKING ON COLUMNS

```
#we are selecting the columns that we are going to use on Telcom_dataset1
df = df[['PRODUCT','VALUE','DATE_TIME','CELL_ID','SITE_ID']]
df
```

```
#we are selecting the columns that we are going to use in the Telcom_dataset2
df2 = df2[['PRODUCT','VALUE','DATE_TIME','CELL_ID','SITE_ID']]
df2
```

```
# we are selecting the columns that we are going to use in the Telcom_dataset3
df3 = df3[["PRODUCT","VALUE","DATE_TIME", "CELL_ID","SITE_ID"]]
df3
```

MERGING OF DATASETS


```
#merging telcom datasets based on #DATE_TIME'
#first telcom1 and telcom2
df = pd.merge(df,df2, how='outer', on=['SITE_ID', 'DATE_TIME', 'PRODUCT', 'VALUE','CELL_ID'])
df
df.head(5)

#secondly, telcom1,2 and telom3
df2 = pd.merge(df2,df3, how='outer', on=['SITE_ID', 'DATE_TIME', 'PRODUCT', 'VALUE','CELL_ID'])
df2
df2.head(5)

#drop duplicates
df2.duplicated()
df_c = df2.drop_duplicates()
df_c

#removing any null records
df_cleaned = df_c.dropna(how='all')
df_cleaned
#df_cleaned.head(2)
```

```
#merge cell_geo and telcom datasets based on SITE_ID

df1_all = pd.merge(df1,df_cleaned, how='inner', on=['SITE_ID'])
df1_all
#df_all.head(2)

#dropping of duplicates based on columns date_time, product and value
df1_all1 = df1_all.drop_duplicates(['DATE_TIME','PRODUCT', 'VALUE'], keep='last')
df1_all1

#resetting the indexes
df1_all1.reset_index(inplace = True) 
df1_all1

#dropping any records with nulls
df1_all2 = df1_all1.dropna(how='all')
df1_all2


#drop unwanted columns
df1_all2.drop(['index', 'CELL_ID_x', 'CELL_ID_y'], axis=1, inplace=True)
df1_all2
#df1_all2
```
